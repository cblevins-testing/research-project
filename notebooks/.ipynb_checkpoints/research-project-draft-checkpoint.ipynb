```python
import pandas as pd
import os
```

```python
#example_df = pd.read_excel("../data/original/example-1.xlsx", sheet_name="Data")
example_df = pd.read_csv("../data/original/damron-gaias-guide-geocoded.csv")
```

```python
example_df.head()
```

```python
import matplotlib.pyplot as plt

# Get the 10 most frequent values in the 'type' column
top_types = example_df['type'].value_counts().head(10)

# Plot the data
top_types.plot(kind='bar', figsize=(10, 6), color='skyblue')
plt.title('Top 10 Most Frequent Types')
plt.xlabel('Type')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```
